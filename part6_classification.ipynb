# AutoML does Classification too!

Here's a prior experiment on classifying Christmas Movies vs Non-Christmas Movies I ran in the Azure Portal to classify Die Hard.

```python
# Load the workspace information from config.json using the Azure ML SDK
from azureml.core import Workspace

ws = Workspace.from_config()
ws.name
```

```python
# List all available primary classification metrics, for reference
from azureml.train.automl.utilities import get_primary_metrics

get_primary_metrics('classification')
```

```python
from azureml.core import Experiment

experiment = Experiment(ws, 'DieHardChristmas')
experiment
```

```python
from azureml.widgets import RunDetails

# Get the most recent run
run = next(experiment.get_runs())
# Or explicitly by ID: run = AutoMLRun(experiment, 'AutoML_e5ca49e8-6904-4f46-aef6-d835d95a1aa7')

# Display details of that run
RunDetails(run).show()
```

```python
# Grab the resulting model and best run
best_auto_run, automl_model = run.get_output()

# Display details about the best run
RunDetails(best_auto_run).show()
```
